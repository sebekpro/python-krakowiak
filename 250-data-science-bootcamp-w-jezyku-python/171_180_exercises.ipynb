### <a name='2'></a> Ćwiczenie 182
Zwróć uwagę, że kolumna _timestamp_ jest typu _object_. Zamień ją na typ _datetime_.

__Wskazówka:__ Użyj funkcji _pd.to_datetime()_.

```
# tutaj wpisz rozwiązanie
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

Do obiektu _df_ dodaj nową kolumnę o nazwie _hour_ i przypisz godzinę z kolumny _timestamp_. Wyświetl kilka początkowych wierszy obiektu _df_.

```
# tutaj wpisz rozwiązanie
df['hour'] = df['timestamp'].dt.hour
df.head()
```

### <a name='3'></a> Ćwiczenie 183
Do obiektu _df_ dodaj nową kolumnę o nazwie _month_ i przypisz miesiąc z kolumny _timestamp_. Wyświetl kilka początkowych wierszy obiektu _df_.

```
# tutaj wpisz rozwiązanie

df['month'] = df['timestamp'].dt.month
df.head()
```

Pogrupuj dane po miesiącu (_month_) oraz policz wartość średnią dla zmiennej _hum_. Wynik przypisz do zmiennej _humidity_by_month_ i zresetuj indeks.

```
# tutaj wpisz rozwiązanie

humidity_by_month = df.groupby('month')['hum'].mean().reset_index()
humidity_by_month
```

Wykorzystując bibliotekę _plotly express_ zbuduj wykres liniowy średniej wartości zmiennej _hum_ dla danego miesiąca (wykorzystaj obiekt _humidity_by_month_).

__Wskazówka:__ Użyj funkcji _px.line()_.

```
# tutaj wpisz rozwiązanie
px.line(humidity_by_month, x = 'month', y = 'hum', title = ' Średnia wartość zmiennej hum dla miesięcy')
```

Zbuduj ten sam wykres tym razem używając ciemnego szablonu biblioteki _plotly express_.

```
# tutaj wpisz rozwiązanie
px.line(humidity_by_month, x = 'month', y = 'hum', title = 'Średnia wartość zmiennej hum dla każdego miesiaca', template = 'plotly_dark')
```

Odczytaj dla jakiego miesiąca średnia wartość zmiennej _hum_ jest najmniejsza i dla jakiego największa.

### <a name='4'></a> Ćwiczenie 184
Pogrupuj dane na poziomie godziny (_hour_) i następnie wyznacz wartość średnią zmiennej _cnt_. Wynik przypisz do zmiennej _cnt_by_hour_ i zresetuj indeks.

```
# tutaj wpisz rozwiązanie
cnt_by_hour = df.groupby('hour')['cnt'].mean().reset_index()
cnt_by_hour
```

Wykorzystując bibliotekę _plotly express_ zbuduj wykres liniowy średniej wartości zmiennej _cnt_ w rozbiciu na godziny (wykorzystaj obiekt _cnt_by_hour_).

```
# tutaj wpisz rozwiązanie
px.line(cnt_by_hour, x = 'hour', y = 'cnt', title = 'Średnia wartość zmiennej cnt w poszczególnych godzinach', labels = {'hour': 'Godzina', 'cnt': 'Średnia cnt'})
```

W powyższym kodzie:

* Używamy funkcji px.line do wygenerowania wykresu liniowego.
* Parametr x określa zmienną na osi poziomej (godziny).
* Parametr y określa zmienną na osi pionowej (średnia wartość cnt).
* Parametr title ustawia tytuł wykresu.
* Parametr labels pozwala na dostosowanie etykiet osi.

### <a name='5'></a> Ćwiczenie 185
Pogrupuj dane po zmiennych _is_weekend_ i _hour_ i wyznacz średnią wartość zmiennej _cnt_. Wynik przypisz do zmiennej _cnt_by_weekend_hour_ i zresetuj indeks.

```
# tutaj wpisz rozwiązanie
cnt_by_weekend_hour = df.groupby(['is_weekend', 'hour'])['cnt'].mean().reset_index()
cnt_by_weekend_hour
```

W powyższym kodzie:

* Używamy metody groupby, aby pogrupować dane według zmiennych is_weekend oraz hour.
* Następnie dla każdej grupy obliczana jest średnia wartość zmiennej cnt.
*Funkcja reset_index() przekształca wynikowy obiekt Series do DataFrame z domyślnym indeksem.

Wykorzystując bibliotekę _plotly express_ zbuduj wykres liniowy średniej wartości zmiennej _cnt_ w rozbiciu na godziny __w dni poza weekendem__ (wykorzystaj obiekt _cnt_by_weekend_hour_). Nadaj tytuł wykresu: 'Poza weekendem'.

```
# tutaj wpisz rozwiązanie

# Filtrowanie danych dla dni poza weekendem (przyjmujemy, że 0 oznacza dni poza weekendem)
cnt_weekdays = cnt_by_weekend_hour[cnt_by_weekend_hour['is_weekend'] == 0]

fig = px.line(cnt_weekdays, x = 'hour', y = 'cnt', title = 'Poza weekendem', labels = {'hour': 'Godzina', 'cnt': 'Średnia cnt'})

fig.show()
```

* Dane są filtrowane po kolumnie is_weekend.
* Następnie funkcja px.line generuje wykres, gdzie oś X reprezentuje godziny, a oś Y – średnią wartość zmiennej cnt.
* Tytuł wykresu został ustawiony na 'Poza weekendem'.

Wykorzystując bibliotekę _plotly express_ zbuduj wykres liniowy średniej wartości zmiennej _cnt_ w rozbiciu na godziny __w dni weekendowe__ (wykorzystaj obiekt _cnt_by_weekend_hour_). Nadaj tytuł wykresu: 'Weekend'.

```
# tutaj wpisz rozwiązanie

# Filtrowanie danych dla dni weekendowych (przyjmujemy, że 1 oznacza weekend)
cnt_weekend = cnt_by_weekend_hour[cnt_by_weekend_hour['is_weekend'] == 1]

# Tworzymy wykres liniowy średniej wartości 'cnt' w zależności od godziny
fig = px.line(cnt_weekend, x = 'hour', y = 'cnt', title = 'Weekend', labels = {'hour': 'Godzina', 'cnt': 'Średnia cnt'})
fig.show()
```

* Dane są filtrowane po kolumnie is_weekend dla wartości 1 (dni weekendowe).
* Funkcja px.line generuje wykres, gdzie oś X reprezentuje godziny, a oś Y – średnią wartość zmiennej cnt.
* Tytuł wykresu został ustawiony na 'Weekend'.
