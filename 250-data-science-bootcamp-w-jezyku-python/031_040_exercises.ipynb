### <a name='1'></a> Ćwiczenie 31
Wykonaj dwa sortowania podanej poniżej tablicy $A$:
* wierszowo (rosnąco)
* kolumnowo (rosnąco)


```
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 4.99],
              [14.99, 2.39, 7.29]])
```



__Wskazówka:__ Użyj funkcji _np.sort()_.

````
# tutaj wpisz rozwiązanie
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 4.99],
              [14.99, 2.39, 7.29]])
ar = np.sort(A, axis = 1)
ac = np.sort(A, axis = 0)

print(f"Sortowanie wierszowo:\n {ar}")
print("\n")

print(f"Sortowanie kolumnowo:\n {ac}")
print("\n")
````

### <a name='3'></a> Ćwiczenie 33
Podana jest poniższa tablica $A$. Zastąp elementy powyżej wartości 10 wartością stałą 10 (obcięcie wartości do 10).


```
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 14.99],
              [14.99, 2.39, 7.29]])
```



__Wskazówka:__ Użyj funkcji _np.where()_.

````
# tutaj wpisz rozwiązanie
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 14.99],
              [14.99, 2.39, 7.29]])
a1 = np.where(A > 10, 10, A)
a1
````
