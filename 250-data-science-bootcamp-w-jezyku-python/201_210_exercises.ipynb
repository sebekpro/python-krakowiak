### <a name='1'></a> Ćwiczenie 201
Z poniższego słownika _data_ utwórz obiekt _DataFrame_ i przypisz do zmiennej _df_.

```
data = {
    'size': ['XL', 'L', 'M', np.nan, 'M', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red', 'green'],
    'gender': ['female', 'male', np.nan, 'female', 'female', 'male'],
    'price': [199.0, 89.0, np.nan, 129.0, 79.0, 89.0],
    'weight': [500, 450, 300, np.nan, 410, np.nan],
    'bought': ['yes', 'no', 'yes', 'no', 'yes', 'no']
}

# tutaj wpisz rozwiązanie
df = pd.DataFrame(data)
df
```

### <a name='3'></a> Ćwiczenie 203
Wykorzystując bibliotekę do uczenia maszynowego _scikit-learn_ oraz klasę _SimpleImputer_ uzupełnij braki danych dla zmiennej _weight_ wartością średnią. Zmiany przypisz na stałe do obiektu _df_.

```
# tutaj wpisz rozwiązanie
from sklearn.impute import SimpleImputer

# Inicjalizacja SimpleImputer z strategią "mean"
imputer = SimpleImputer(missing_values=np.nan, strategy = 'mean')

# Uzupełnienie brakujących wartości w kolumnie 'weight' wartością średnią
df[['weight']] = imputer.fit_transform(df[['weight']])

df
```

1. **Inicjalizacja SimpleImputer**:
Tworzymy obiekt imputer z parametrem strategy="mean", co oznacza, że brakujące wartości zostaną uzupełnione średnią wartością z danej kolumny.

2. **Transformacja**:
Metoda fit_transform() jest stosowana do kolumny weight. Wynik (macierz z uzupełnionymi wartościami) przypisujemy z powrotem do df[['weight']].

Wyświetl wartość średnią wstawioną w miejsce braków kolumny _weight_.

```
# tutaj wpisz rozwiązanie
imputer.statistics_
```

### <a name='4'></a> Ćwiczenie 204
Wykorzystując bibliotekę do uczenia maszynowego _scikit-learn_ oraz klasę _SimpleImputer_ uzupełnij braki danych dla zmiennej _price_ stałą wartością 99.0. Zmiany przypisz na stałe do obiektu _df_.

```
# tutaj wpisz rozwiązanie

# Inicjalizacja SimpleImputer z strategią "constant" i wartością fill_value=99.0
imputer_price = SimpleImputer(strategy = 'constant', fill_value = 99.0)

# Uzupełnienie brakujących wartości w kolumnie 'price'
df[['price']] = imputer_price.fit_transform(df[['price']])

df
```

1. **Inicjalizacja SimpleImputer**:
Obiekt imputer_price jest tworzony z parametrem strategy="constant", co powoduje, że braki danych są zastępowane wartością podaną w fill_value (czyli 99.0).

2. **Transformacja**:
Metoda fit_transform() jest wywoływana na kolumnie price, uzupełniając wszystkie wartości NaN stałą wartością 99.0. Wynik przypisywany jest bezpośrednio do oryginalnej kolumny price w obiekcie df.

### <a name='9'></a> Ćwiczenie 209
Usuń wiersze z brakami danych z obiektu _df_ dla których nie ma co najmniej 5 zdefiniowanych wartości (inaczej zawierają dwa braki _np.nan_). Zmian nie przypisuj na stałe do zmiennej _df_.

```
# tutaj wpisz rozwiązanie
df_drop = df.dropna(thresh = 5)
df_drop
```

1. Metoda dropna() z parametrem thresh:
Wywołanie df.dropna(thresh=5) powoduje, że zostaną zachowane tylko te wiersze, które mają przynajmniej 5 wartości nie będących NaN.

* Parametr thresh=5 ustala minimalną liczbę niepustych wartości, jakie dany wiersz musi posiadać, aby nie został usunięty.
* Wiersze, które mają mniej niż 5 zdefiniowanych wartości (czyli zawierają co najmniej 2 braki), zostaną wycięte z wyniku.

### <a name='10'></a> Ćwiczenie 210
Usuń wiersze z brakami danych z obiektu _df_ dla których nie ma co najmniej 5 zdefiniowanych wartości (inaczej zawierają dwa braki _np.nan_) a resztę braków wypełnij stałą wartością 400.0. Zmian nie przypisuj na stałe do zmiennej _df_.

```
# tutaj wpisz rozwiązanie

# Najpierw usuwamy wiersze, które mają mniej niż 5 zdefiniowanych wartości
df_drop = df.dropna(thresh = 5)

# Następnie uzupełniamy pozostałe braki wartością 400.0
df_result = df_drop.fillna(400.0)

df_result
```
