### <a name='6'></a> Ćwiczenie 186
Wykorzystując bibliotekę _plotly express_ zbuduj histogram zmiennej _charges_.

Użyj funkcji _px.histogram()_ z następującymi parametrami:
* width=800
* height=400
* nbins=50
* facet_col='smoker'
* facet_row='sex'

```
# tutaj wpisz rozwiązanie
px.histogram(df, x = 'charges',
             width = 800,           # Szerokość wykresu ustawiona na 800 pikseli
             height = 400,          # Wysokość wykresu ustawiona na 400 pikseli
             nbins = 50,            # Liczba przedziałów (koszyków) histogramu wynosi 50
             facet_col = 'smoker',  # Histogram podzielony na kolumny według zmiennej 'smoker'
             facet_row = 'sex',     # Histogram podzielony na wiersze według zmiennej 'sex'
             title = 'Histogram zmiennej charges')
```

* Parametry width i height: Ustawiają rozmiary wykresu na 800 pikseli szerokości i 400 pikseli wysokości.
* Parametr nbins: Definiuje liczbę przedziałów (koszyków) histogramu, co wpływa na szczegółowość rozkładu.
* Parametry facet_col oraz facet_row: Umożliwiają podział wykresu na mniejsze wykresy (facety) według zmiennych smoker (w kolumnach) oraz sex (w wierszach). Dzięki temu możemy zobaczyć rozkład zmiennej charges w podziale na kategorie palenia i płci.
* Tytuł wykresu: Parametr title ustawia tytuł wykresu na "Histogram zmiennej charges".

Wykorzystując bibliotekę _plotly express_ znajdź rozkład zmiennej _smoker_.

Użyj funkcji _px.histogram()_ z następującymi parametrami:
* width=800
* height=400
* facet_col='sex'
* color='sex'

```
# tutaj wpisz rozwiązanie
fig = px.histogram(df, x = 'smoker',
                   width = 800,             # Szerokość wykresu ustawiona na 800 pikseli
                   height = 400,            # Wysokość wykresu ustawiona na 400 pikseli
                   facet_col = 'sex',       # Podział wykresu na kolumny według zmiennej 'sex'
                   color = 'sex',           # Kolorowanie danych według zmiennej 'sex'
                   title = 'Histogram zmiennej smoker'
                   )
fig.show()
```

* Funkcja px.histogram(): Generuje histogram dla zmiennej smoker z obiektu DataFrame df.
* Parametr x='smoker': Określa, że oś X wykresu reprezentuje zmienną smoker.
* Parametry width i height: Ustawiają rozmiar wykresu na 800 pikseli szerokości i 400 pikseli wysokości.
* Parametr facet_col='sex': Dzieli wykres na osobne kolumny w zależności od wartości zmiennej sex.
* Parametr color='sex': Wykorzystuje zmienną sex do kolorowania wykresu, co umożliwia lepsze porównanie rozkładu w poszczególnych grupach.
* Tytuł wykresu: Ustawia tytuł na "Rozkład zmiennej smoker".
* fig.show(): Wyświetla interaktywny wykres w środowisku obsługującym Plotly.

### <a name='7'></a> Ćwiczenie 187
Skonwertuj zmienne kategoryczne obietku _df_ do kodowania 0-1 i przypisz do zmiennej _df_dummies_.

__Wskazówka:__ Użyj funkcji _pd.get_dummies()_ z parametrem _drop_first=True_.

```
# tutaj wpisz rozwiązanie
df_dummies = pd.get_dummies(df,drop_first = True)
print(df.info())
```

* pd.get_dummies(df, drop_first=True): Funkcja ta konwertuje wszystkie zmienne kategoryczne w obiekcie df na kolumny binarne (0 lub 1).
* drop_first=True: Ustawienie tego parametru powoduje, że z każdej zmiennej kategorycznej pomijana jest pierwsza kategoria, co redukuje nadmiarowość danych.
* Wynikowa DataFrame jest przypisana do zmiennej df_dummies.

Funkcja **pd.get_dummies()** przekształca kolumny kategoryczne w DataFrame na zestaw kolumn zero-jedynkowych (dummy variables), gdzie każda nowa kolumna reprezentuje jedną z kategorii danej zmiennej. Używając parametru **drop_first=True**, dla każdej zmiennej kategorycznej zostaje pominięta pierwsza kategoria, aby uniknąć redundancji (tzw. dummy variable trap).

W przypadku:

* **sex**: Oryginalna zmienna miała dwie kategorie (np. "female" i "male"). Po kodowaniu, jedna z nich (np. "female") jest pominięta, a w rezultacie powstaje kolumna **sex_male**. Wartość **True** w tej kolumnie oznacza, że dany wiersz dotyczy osoby płci męskiej.
* **smoker**: Podobnie, dla zmiennej **smoker** z kategoriami (np. "no" i "yes") zostaje utworzona tylko jedna kolumna, czyli **smoker_yes**. Wartość **True** wskazuje, że osoba pali.
* **region**: Jeśli zmienna **region** miała cztery kategorie (np. "northeast", "northwest", "southeast", "southwest"), to przy pominięciu pierwszej kategorii (najprawdopodobniej "northeast") powstaną trzy nowe kolumny: region_northwest, **region_southeast** i **region_southwest**. Każda z nich przyjmuje wartość **True** dla wierszy, które odpowiadają danej kategorii regionu.

W rezultacie, poza oryginalnymi zmiennymi numerycznymi, otrzymujesz 5 dodatkowych kolumn reprezentujących zakodowane zmienne kategoryczne.

### <a name='8'></a> Ćwiczenie 188
Wyznacz korelacje zmiennych w obiekcie _df_dummies_ i przypisz do zmiennej _corr_.

```
# tutaj wpisz rozwiązanie
corr = df_dummies.corr()
corr
```

* df_dummies.corr(): Ta metoda oblicza macierz korelacji Pearsona dla wszystkich kolumn w DataFrame df_dummies.
* Wynikowa macierz, która zawiera współczynniki korelacji między zmiennymi, jest przypisywana do zmiennej corr.

Wytnij kolumnę z obiektu _corr_ dla zmiennej _charges_ (jako obiekt DataFrame) i posortuj wartości malejąco.

```
# tutaj wpisz rozwiązanie
corr_charges = corr[['charges']].sort_values(by = 'charges', ascending = False)
corr_charges
```

* corr[['charges']] – Wycinamy kolumnę charges z obiektu corr. Użycie podwójnych nawiasów zwraca wynik jako DataFrame.
* .sort_values(by='charges', ascending=False) – Sortujemy wartości w kolumnie charges w porządku malejącym.


Wykorzystując bibliotekę _plotly express_ utwórz wykres słupkowy dla tak posortowanych wartości. Pomiń pierwszy element obiektu.

```
# tutaj wpisz rozwiązanie
filtered_corr = corr_charges.iloc[1:]
fig = px.bar(filtered_corr, x = filtered_corr.index, y = 'charges',
             title = 'Korelacje zmiennych z charges (pomijając pierwszy element)',
             labels = {'index': 'Zmienne', 'charges': 'Współczynnik korelacji'})
fig.show()
```

* filtered_corr = corr_charges.iloc[1:]: Ta linia usuwa pierwszy wiersz z DataFrame, który odpowiada korelacji zmiennej charges z samą sobą.
* px.bar(...): Funkcja tworzy wykres słupkowy, gdzie:
x=filtered_corr.index ustawia etykiety słupków (nazwy zmiennych).
y='charges' określa wartości na osi Y (współczynniki korelacji z charges).
* fig.show(): Wyświetla interaktywny wykres.

Wykorzystując bibliotekę _scikit-learn_ oraz funkcję _train_test_split_ podziel dane z obiektów _data_ i _target_ na zbiór treningowy (_X\_train_, _y\_train_) oraz zbiór testowy (_X\_test_, _y\_test_). Do zbioru testowego przydziel 20% próbek.

Wyświetl rozmiar obiektów:
* _X\_train_
* _X\_test_
* _y\_train_
* _y\_test_

```
# tutaj wpisz rozwiązanie
from sklearn.model_selection import train_test_split

# Podział danych na zbiór treningowy i testowy (20% próbek w zbiorze testowym)
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.20, random_state = 42)

# Wyświetlenie rozmiaru obiektów
print("Rozmiar X_train", X_train.shape)
print("Rozmiar X_test", X_test.shape)
print("Rozmiar y_train", y_train.shape)
print("Rozmiar y_test", y_test.shape)
```

* Importujemy funkcję train_test_split z modułu sklearn.model_selection.
* Funkcja train_test_split dzieli dane na zbiory treningowy i testowy. Parametr test_size=0.20 oznacza, że 20% danych trafia do zbioru testowego.
* random_state=42 ustawia ziarno generatora liczb losowych, aby wynik podziału był powtarzalny.
* Po podziale wyświetlamy kształty (rozmiary) obiektów: X_train, X_test, y_train oraz y_test.

Wybierając wartość **42**, gwarantujemy, że podział danych na zbiory treningowy i testowy będzie taki sam przy każdym wykonaniu, co ułatwia porównywanie eksperymentów i debugowanie. Sam numer **42** nie ma specjalnego znaczenia technicznego – jest często wybierany ze względu na kultowy odniesienie do "odpowiedzi na pytanie o życie, wszechświat i całą resztę" z książki Douglasa Adamsa, ale pod kątem funkcjonalnym mógłby to być dowolny inny numer całkowity.

Wynikowe wymiary oznaczają, że:

* X_train (1069, 8): Zbiór treningowy zawiera 1069 próbek (wierszy) oraz 8 cech (kolumn). Każda próbka reprezentuje obserwację (np. osobę, przypadek) opisaną 8 cechami.
* X_test (268, 8): Zbiór testowy zawiera 268 próbek z tymi samymi 8 cechami. Te dane są używane do oceny modelu po jego wytrenowaniu.
* y_train (1069,): Zbiór etykiet (target) dla danych treningowych składa się z 1069 wartości, czyli jednej wartości dla każdej próbki w zbiorze treningowym.
* y_test (268,): Zbiór etykiet dla danych testowych składa się z 268 wartości.

Oznacza to, że dane zostały podzielone na 80% próbek (1069) przeznaczonych do trenowania modelu oraz 20% próbek (268) do jego testowania, przy czym każdy przykład w zbiorze cech odpowiada jednej wartości docelowej (target).
