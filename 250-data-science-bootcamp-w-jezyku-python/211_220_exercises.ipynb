### <a name='1'></a> Ćwiczenie 211
Podany jest poniższy obiekt _DataFrame_:

```
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
```
Dokonaj dyskretyzacji zmiennej _height_ na 3 przedziały o równej szerokości. Wynik przypisz do nowej kolumny _height_cut_.


````
# tutaj wpisz rozwiązanie
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})

# Dyskretyzacja zmiennej 'height' na 3 przedziały o równej szerokości
df['height_cut'] = pd.cut(df['height'], bins = 3)

df
````

Funkcja pd.cut dzieli ciągły zakres wartości zmiennej na określoną liczbę przedziałów o równej szerokości. W przypadku zmiennej height oraz argumentu bins=3:

1. Obliczanie zakresu wartości:
Funkcja wyznacza minimalną i maksymalną wartość w kolumnie height. Dla przykładu, jeśli mamy wartości od 168.0 do 191.0, to zakres wynosi 23.0.

2. Wyznaczenie szerokości przedziału:
Całkowity zakres jest dzielony przez liczbę przedziałów. W naszym przykładzie szerokość jednego przedziału wynosi 23.0/0 ≈ 7.67

3. Ustalenie granic przedziałów:
Na podstawie minimalnej wartości oraz szerokości przedziału funkcja automatycznie określa granice:
* Pierwszy przedział: od 168.0 do 168.0 + 7.67 ≈ 175.67,
* Drugi przedział: od 175.67 do 175.67 + 7.67 ≈ 183.33,
* Trzeci przedział: od 183.33 do 191.0.

Domyślnie przedziały są tworzone jako otwarte z lewej i domknięte z prawej stroną, czyli np. (168.0, 175.67].

4. Przypisanie wartości do przedziałów:
Każda wartość w kolumnie height jest przyporządkowywana do odpowiedniego przedziału, w którym się znajduje. W wyniku tej operacji otrzymujemy zmienną kategoryczną, gdzie każda obserwacja ma etykietę przedziału (np. (175.67, 183.33]).

Podsumowując, dyskretyzacja zmiennej height przy użyciu pd.cut(height, bins=3) polega na automatycznym podziale zakresu wartości na 3 równe segmenty, a następnie przypisaniu każdej wartości do odpowiadającego jej segmentu. Wynik jest nową kolumną, która kategoryzuje wartości zmiennej height według ustalonych przedziałów.
